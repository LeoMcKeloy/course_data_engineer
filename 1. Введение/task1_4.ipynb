**1. Построение распределений**

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from IPython import display
```

```python
x = np.linspace(sps.rayleigh.ppf(0.01), sps.rayleigh.ppf(0.99), 100)
plt.hist(sps.rayleigh.pdf(x), label='rayleigh pdf')
plt.legend()
plt.show()
```

**2. Построение графика на по функции**

```python
plt.title('(x2+y2−1)3−x2y3=0', fontsize=16)
x = np.linspace(-1, 1, 200)
y1 = np.sqrt(1-np.power(x, 2)) + np.power(np.square(x), 0.33)
y2 = -np.sqrt(1-np.power(x, 2)) + np.power(np.square(x), 0.33)
my_x_ticks = np.arange(-1.5, 2, 0.5)
my_y_ticks = np.arange(-1.5, 2, 0.5)
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='r')
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)
plt.grid()
plt.show()
```

**3. Анализ зависимостей**

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

with open("./dataset_home.txt", mode="r") as file:
  for human in file.readlines():
    age.append(int(human.split(",")[0]))

    if human.split(",")[1] == "male":
      sex.append(1)
    else:
      sex.append(0)
    
    bmi.append(float(human.split(",")[2]))

    children.append(int(human.split(",")[3]))

    if human.split(",")[4] == "yes":
      smoker.append(1)
    else:
      smoker.append(0)
    
    if human.split(",")[5] == "northwest":
      region.append(0)
    elif human.split(",")[5] == "northeast":
      region.append(1)
    elif human.split(",")[5] == "southwest":
      region.append(2)
    elif human.split(",")[5] == "southeast":
      region.append(3)
    
    charges.append(float(human.split(",")[6]))
```

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charges = np.array(charges)
```

```python
plt.scatter(charges[age<36], bmi[age<36], label = "x < 36")
plt.scatter(charges[age>35][age[age>35]<56], bmi[age>35][age[age>35]<56], label = "36 < x < 56")
plt.scatter(charges[age>55], bmi[age>55], label = "x > 55")
plt.legend()
plt.show()
```
